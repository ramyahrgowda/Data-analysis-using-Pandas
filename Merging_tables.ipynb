```python
import pandas as pd
import numpy as np
```

```python
df_churn_raw = pd.read_csv('Churn_in_telecom_dataset.csv')
```

### concat, merge,

```python
df_churn_raw.shape
```

```python
df_churn_raw.columns
```

```python
#Dividing data into 2 dataframes
df_churn_1 = df_churn_raw.iloc[0:1500,:]
df_churn_2 = df_churn_raw.iloc[1500:,:]
```

```python
df_churn_1.shape
```

```python
df_churn_2.shape
```

```python
#splitting data column wise
df_churn_day = df_churn_raw[['phone number','total day minutes', 'total day calls', 'total day charge']]
df_churn_eve = df_churn_raw[['phone number','total eve minutes', 'total eve calls', 'total eve charge']]
df_churn_night = df_churn_raw[['total night minutes', 'total night calls', 'total night charge']]
```

#### Concat

```python
#combining data split row-wise
pd.concat([df_churn_1,df_churn_2])
```

```python
#combining data split row-wise
df_churn_1.append(df_churn_2)
```

#### merge

```python
#merging data - left join on column 'phone number'
pd.merge(left=df_churn_day,right=df_churn_eve,how='left',left_on='phone number',right_on='phone number')
```

```python
#concat can also be used if index are matching
pd.concat([df_churn_day,df_churn_eve,df_churn_night],axis=1)
```
